```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import logm, expm

def get_similitude_matrix(scale, theta_deg, tx, ty):
    """Similitude Matrix """
    theta = np.radians(theta_deg)
    c, s = np.cos(theta), np.sin(theta)
    A = np.eye(3)
    A[0, 0], A[0, 1], A[0, 2] = scale * c, -scale * s, tx
    A[1, 0], A[1, 1], A[1, 2] = scale * s,  scale * c, ty
    return A

def transform_shape(shape, A):
    """  """
    ones = np.ones((shape.shape[0], 1))
    points_h = np.hstack([shape, ones])
    transformed_h = (A @ points_h.T).T
    return transformed_h[:, :2]

# --- 1.   ( ) ---
base_shape = np.array([[0, 0], [2, 0], [2, 2], [1, 3], [0, 2], [0, 0]])
base_shape = base_shape - np.mean(base_shape, axis=0) #  

# --- 2.     (Start & End) ---
# (scale, rotation_deg, tx, ty)
start_params = (1.0, 0, 2, 2)
end_params   = (0.5, 90, 8, 8)

A_start = get_similitude_matrix(*start_params)
A_end   = get_similitude_matrix(*end_params)

# --- 3. Logarithm Map (Lie Group -> Lie Algebra) ---
L_start = logm(A_start)
L_end   = logm(A_end)

# --- 4.     ---
plt.figure(figsize=(10, 8))
plt.title("Part (1): Linear Interpolation between Two Deformations")
plt.axis('equal')
plt.grid(True, linestyle='--', alpha=0.5)

#  
steps = 20
cmap = plt.cm.viridis

for i, t in enumerate(np.linspace(0, 1, steps)):
    # 4-1. Lie Algebra   
    # L(t) = (1-t) * L_start + t * L_end
    L_t = (1 - t) * L_start + t * L_end
    
    # 4-2. Exponential Map (Lie Algebra -> Lie Group)
    A_t = expm(L_t)
    
    # 4-3.    
    deformed_shape = transform_shape(base_shape, A_t)
    draw_shape = np.vstack([deformed_shape, deformed_shape[0]]) # 
    
    #   ,  
    if i == 0 or i == steps - 1:
        plt.plot(draw_shape[:, 0], draw_shape[:, 1], 'k-', linewidth=2, label="Keyframe" if i==0 else "")
    else:
        color = cmap(t)
        plt.plot(draw_shape[:, 0], draw_shape[:, 1], color=color, alpha=0.6)

plt.legend()
plt.savefig("output/extra_credit_part1.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm

def get_similitude_matrix(scale, theta_deg, tx, ty):
    """
    Construct a 2D similitude matrix

    A similitude matrix in SIM(2) is an affine transformation consisting of 
    a uniform magnification, a rotation by a specific angle, and a translation vector
    In other words, from the professor's note, SIM(2) encodes all transformations that preserve shape 
    but change size (magnification), orientation, and position.

    Mathmatically, a similitude is represented as
    A=[[a*R(theta) | t],
        [theta | 1]], where
        R(theta) = [[cos(theta), -sin(theta)],
                    [sin(theta), cos(theta)]], and
        t = (tx, ty)^T

    A point (x, y) in 2D
    Thus, (x', y') a*R(theta)(x,y)+t.
    This means that magnification by a, rotation by theta, translation by (tx, ty)

    Return will be SIM(2) matrix.
    SIM(2) matrix A is 
    [[a*cos(theta), -a*sin(theta), tx],
     [a*sin(theta), a*cos(theta), ty],
     [0, 0, 1]]
    """
    theta=np.radians(theta_deg)
    c, s=np.cos(theta), np.sin(theta) # cos(theta), sin(theta)
    A = np.eye(3) # 3x3 matrix

    # Magnification and rotation
    A[0, 0], A[0, 1] = scale* c, -scale* s
    A[1, 0], A[1, 1] = scale* s,  scale* c

    # Translation
    A[0, 2], A[1,2]= tx, ty
    return A

def similitude_log(A):
    """
    Compute the logarithm of a SIM(2) matrix and return its corresponding element in sim(2)
    A= [[a*cos(theta), -a*sin(theta), tx],
         [a*sin(theta), a*cos(theta), ty],
         [0, 0, 1]]
    
    The logarithm(A) lies in the sim(2), consisting of matrices
    B= [[lambda, -theta, u],
        [theta, lambda, v],
        [theta, theta, theta]], where lambda is natural log (a),
    and (u, v) encodes the infinitesimal translation after converting through the matrix V defined in B1.
    V=Omega^{-1}*(e^{Omega}-I), where Omega= [[lambda, -theta],
                                                [theta, lambda]]
    """
    # Extract magnification  from the 22 rotational block
    alpha = np.sqrt(A[0,0]**2 + A[0,1]**2)
    
    # lambda = ln(a)
    lam = np.log(alpha)
    
    # Extract rotation angle 
    theta = np.arctan2(A[1, 0], A[0, 0])
    
    # Form Omega
    Omega = np.array([[lam, -theta],
                      [theta, lam]])
    
    # Extract translation vector W
    W = np.array([A[0, 2], A[1, 2]])
    
    # Compute V using the expression from Problem B1 part e
    if abs(lam) < 1e-10 and abs(theta) < 1e-10:
        # Case lambda = rotation angle = 0, so V = I
        V = np.eye(2)
    else:
        # V
        exp_Omega = np.array([[alpha * np.cos(theta), -alpha * np.sin(theta)],
                              [alpha * np.sin(theta),  alpha * np.cos(theta)]])
        V = np.linalg.solve(Omega, exp_Omega - np.eye(2))
    
    # Compute for U from W=VU
    U = np.linalg.solve(V, W)
    
    # Assemble the sim(2) matrix B.
    B = np.zeros((3, 3))
    B[0, 0], B[0, 1], B[0, 2] = lam, -theta, U[0]
    B[1, 0], B[1, 1], B[1, 2] = theta, lam, U[1]
    
    return B

def transform_shape(shape, A):
    """
    Apply a SIM(2) matrix to a 2D shape.

    A 2D point (x, y) is first lifted into coordinates as (x, y, 1).
    And then, (x', y', 1)^T=A*(x, y, 1)^T, where A is SIM(2) matrix

    Therefore, the transformed point is (x', y')=a*R(theta)(x, y)+t.

    This function performs the transformation on an entire set of points. 
    """
    
    ones = np.ones((shape.shape[0], 1)) # Convert points (x, y, 1) form
    points_h = np.hstack([shape, ones]) 

    # Apply the transformation A*(x, y, 1)^T
    transformed_h = (A @ points_h.T).T 

    return transformed_h[:, :2] # Return should be the (x',y') part

# Base shape
# We choose a 2D pentagon shape.
base_shape = np.array([[0, 0], [2, 0], [2, 2], [1, 3], [0, 2], [0, 0]])
base_shape = base_shape - np.mean(base_shape, axis=0)

# Define two endpoint similitudes
start_params = (1.0, 0, 2, 2) # a, rotation degree, translation (2, 2)
end_params   = (0.5, 90, 8, 8) # a, rotation degree, translation (8, 8)

A_start = get_similitude_matrix(*start_params)
A_end   = get_similitude_matrix(*end_params)

# Compute log (SIM(2) -> sim(2))
L_start = similitude_log(A_start)
L_end   = similitude_log(A_end)

#check
print("Start: ", L_start)
print("\nEnd: ", L_end)

# Linear interpolation in sim(2) and visualization
plt.figure(figsize=(10, 8))
plt.title("B2 Part (1): Linear interpolation between two deformations")
plt.axis('equal')
plt.grid(True, linestyle='--', alpha=0.5)

steps = 20
cmap = plt.cm.viridis

for i, t in enumerate(np.linspace(0, 1, steps)):
    # Start interpolation in sim(2)
    L_t = (1 - t) * L_start + t * L_end
    
    # Exponential map back to SIM(2)
    A_t = expm(L_t)
    
    # Apply the interpolated transformation to the shape
    deformed_shape = transform_shape(base_shape, A_t)
    
    # Plot
    if i == 0 or i == steps - 1:
        plt.plot(deformed_shape[:, 0], deformed_shape[:, 1], 'k-', 
                linewidth=2.5, marker='o', markersize=4,
                label="Keyframe" if i == 0 else "")
    else:
        color = cmap(t)
        plt.plot(deformed_shape[:, 0], deformed_shape[:, 1], 
                color=color, alpha=0.6, linewidth=1)

plt.legend()
plt.savefig("output/b2_part1.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from scipy.linalg import expm # Reverse map (exp) is still needed from scipy
from interpatxy import interpatxy  # Project 3 function

# Import the manual functions defined in Part 1
from b2_part1 import get_similitude_matrix, similitude_log, transform_shape

def eval_cubic_bezier(t, P0, P1, P2, P3):
    return (1-t)**3 * P0 + 3*(1-t)**2*t * P1 + 3*(1-t)*t**2 * P2 + t**3 * P3

# --- 1. Initial Setup (House Shape) ---
base_shape = np.array([[0, 0], [2, 0], [2, 2], [1, 3], [0, 2], [0, 0]])
base_shape = base_shape - np.mean(base_shape, axis=0)

# --- 2. Define Deformation Sequence (A0 ... Am) ---
# (scale, rotation_deg, tx, ty)
keyframes_params = [
    (1.0, 0,   2, 2),    # A0: Start
    (0.5, 90,  8, 5),    # A1: 90 deg, shrink, move
    (1.5, 180, 12, 10),  # A2: 180 deg, grow, move
    (1.0, 270, 6, 15),   # A3: 270 deg, restore size, move
    (0.8, 360, 2, 2)     # A4: 360 deg, slightly smaller, return
]

# Use the function imported from Part 1
matrices = [get_similitude_matrix(*p) for p in keyframes_params]
N_frames = len(matrices)

# --- 3. Logarithm Map & Component Extraction ---
# Use the manual logic function imported from Part 1
print("Computing Logarithms using manual similitude_log from Part 1...")
log_matrices = [similitude_log(M) for M in matrices]

# Extract components from the Lie Algebra matrices (sim(2))
# B = [[lambda, -theta, u], [theta, lambda, v], ...]
alphas = np.array([L[0, 0] for L in log_matrices]) # lambda
thetas = np.array([L[1, 0] for L in log_matrices]) # theta
us     = np.array([L[0, 2] for L in log_matrices]) # u
vs     = np.array([L[1, 2] for L in log_matrices]) # v

# --- 4. Compute Spline Control Points using interpnatxy ---
print("Computing Spline Control Points using Project 3 code...")
# We use interpnatxy to get Bezier control points for the parameters in Lie Algebra
_, _, B_alpha, B_theta = interpnatxy(alphas, thetas) 
_, _, B_u, B_v         = interpnatxy(us, vs)
plt.close('all') 

# --- 5. Animation Setup ---
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_title("Part (2): Motion Interpolation (Similitude Spline)")
ax.set_xlim(-2, 16) 
ax.set_ylim(-2, 18)
ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.5)

# Static Elements (Keyframes)
for i, M in enumerate(matrices):
    deformed = transform_shape(base_shape, M)
    draw_s = np.vstack([deformed, deformed[0]])
    ax.plot(draw_s[:, 0], draw_s[:, 1], 'k--', linewidth=1, alpha=0.3)
    ax.text(deformed[0, 0], deformed[0, 1], f"A{i}", fontsize=10, alpha=0.5)

# Dynamic Elements (Moving Body)
line, = ax.plot([], [], 'r-', linewidth=3, label='Moving Body')
path_x, path_y = [], []
trace, = ax.plot([], [], 'b:', linewidth=1, alpha=0.5, label='Trace') 

# Animation Settings
fps = 30
duration_sec = 5
total_frames = fps * duration_sec
segments = N_frames - 1

def update(frame):
    global_t = frame / (total_frames - 1)
    
    # Map global time to segment index (i) and local time (t)
    u = global_t * segments
    i = int(u)
    if i >= segments:
        i = segments - 1
        t = 1.0
    else:
        t = u - i
        
    # 5-1. Spline Evaluation (in Lie Algebra)
    a_t = eval_cubic_bezier(t, *B_alpha[i])
    w_t = eval_cubic_bezier(t, *B_theta[i])
    u_t = eval_cubic_bezier(t, *B_u[i])
    v_t = eval_cubic_bezier(t, *B_v[i])
    
    # 5-2. Exp Map (Lie Algebra -> Lie Group)
    # Reconstruct the sim(2) matrix manually
    L_t = np.zeros((3, 3))
    L_t[0, 0], L_t[0, 1], L_t[0, 2] = a_t, -w_t, u_t
    L_t[1, 0], L_t[1, 1], L_t[1, 2] = w_t,  a_t, v_t
    
    # Use scipy's expm for the reverse mapping (sim(2) -> SIM(2))
    A_t = expm(L_t)
    
    # 5-3. Shape Transformation using imported helper
    deformed = transform_shape(base_shape, A_t)
    draw_s = np.vstack([deformed, deformed[0]])
    
    # Update Plot
    line.set_data(draw_s[:, 0], draw_s[:, 1])
    
    # Update Trace
    center = np.mean(deformed, axis=0)
    path_x.append(center[0])
    path_y.append(center[1])
    trace.set_data(path_x, path_y)
    
    return line, trace

# Create Animation
anim = FuncAnimation(fig, update, frames=total_frames, interval=1000/fps, blit=True)

# Save
save_path = "output/b2_part2.gif"
print(f"Saving animation to {save_path}... (This may take a moment)")
anim.save(save_path, writer=PillowWriter(fps=fps))
print("Done!")

plt.legend()
plt.show()
```
